```python
# using pandas dataframe
batsman = pd.read_csv('/sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma Vs Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dotted',linewidth= 5,marker = 'D',markersize=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue',linestyle = 'solid',linewidth = 4,marker = '+',markersize=3)
plt.title('Rohit Sharma Vs Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# legend
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dotted',linewidth= 5,marker = 'D',markersize=4,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue',linestyle = 'dashed',linewidth = 4,marker = '+',markersize=3,label= 'Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc= 'right')
plt.grid()
plt.show()
```

## Subplots

```python
batter = pd.read_csv('/batter.csv')
batter.head()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_title('Average Vs Strike Rate')
ax.set_xlabel('avg')
ax.set_ylabel('strike rate')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'],color='red')
ax[0].set_title('Avg Vs strike rate')
ax[0].set_ylabel('strike_rate')

ax[1].set_title('AVG vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('avg')
```

## 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
xx
```

```python
yy
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)+ np.log(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## countour plot

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(10,10))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(10,10))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
